````python
from bs4 import BeautifulSoup

# Load your HTML file
with open('players.html', 'r', encoding='utf-8') as f:
    content = f.read()

# Parse HTML content
soup = BeautifulSoup(content, 'html.parser')

# Initialize lists to store data
players = []
categories = []
prices = []

# Find the section that contains player data
for player_row in soup.find_all('div', class_='AUipl_tbl-tr'):  # Replace with actual class
    # Extract player name
    player_name = player_row.find('div', class_='AUipl_tbl-nm AUipl_tbl-PlIc').text.strip()  # Replace with actual class
    players.append(player_name)

    # Extract category
    player_category = player_row.find('div', class_='AUipl_tbl-MTy').text.strip()  # Replace with actual class
    categories.append(player_category)

    # Extract price
    player_price = player_row.find('div', class_='AUipl_tx-CtRt').text.strip()  # Replace with actual class
    prices.append(player_price)

# Print the extracted data
for player, category, price in zip(players, categories, prices):
    print(f'Player: {player}, Category: {category}, Price: {price}')
````

````python
import csv
from bs4 import BeautifulSoup

# Load HTML content
with open('players.html', 'r', encoding='utf-8') as f:
    content = f.read()

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(content, 'html.parser')

# Open a CSV file to write the extracted data
with open('ipl_players.csv', 'w', newline='', encoding='utf-8') as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerow(['Player Name', 'Category', 'Price (₹ Cr.)'])  # Write CSV header

    # Find all the relevant table rows containing player info
    for row in soup.find_all('tr', class_='AUipl_tbl-tr'):
        # Extract the player's name
        player_name_tag = row.find('a')
        player_name = player_name_tag.text.strip() if player_name_tag else 'N/A'

        # Extract the category
        category_tag = row.find('td', class_='AUipl_tx-ctlt')
        category = category_tag.text.strip() if category_tag else 'N/A'

        # Extract the price
        price_tag = row.find('td', class_='AUipl_tx-act')
        price = price_tag.text.strip() if price_tag else 'N/A'

        csvwriter.writerow([player_name, category, price])  # Write to CSV file

print("Data has been successfully written to 'ipl_players.csv'.")
````

````python
import csv

# Read the CSV file and filter out rows with "N/A"
with open('ipl_players.csv', 'r', encoding='utf-8') as infile, open('filtered_ipl_players.csv', 'w', newline='', encoding='utf-8') as outfile:
    csvreader = csv.reader(infile)
    csvwriter = csv.writer(outfile)
    
    # Write the header to the new CSV file
    header = next(csvreader)
    csvwriter.writerow(header)
    
    # Filter and write rows that do not contain "N/A"
    for row in csvreader:
        if "N/A" not in row:
            csvwriter.writerow(row)

print("Filtered data has been successfully written to 'filtered_ipl_players.csv'.")
````

````python
import pandas as pd

# Load the CSV file while keeping lines that have missing values in other columns
data = pd.read_csv('unsold_players.csv', header=0, skip_blank_lines=False)

# Create an empty list to store cleaned rows
cleaned_data = []

# Iterate through the rows
for i in range(len(data)):
    # If the 'Player Name' column is not NaN, it's a proper row with player details
    if pd.notna(data.loc[i, 'Player Name']):
        player_row = data.loc[i]  # Save the current player row
        
        # If the next row is the price (and the current row has an empty price), fix it
        if pd.isna(player_row['Price (₹ Cr.)']) and pd.notna(data.loc[i+1, 'Player Name']) == False:
            player_row['Price (₹ Cr.)'] = data.loc[i+1, 'Player Name']  # The next row contains the price
            cleaned_data.append(player_row)
    else:
        continue

# Convert the cleaned data list back to a DataFrame
cleaned_df = pd.DataFrame(cleaned_data)

# Drop the 'S.No.' column
cleaned_df = cleaned_df.drop(columns=['S.No.'])

# Save the cleaned data to a new CSV file
cleaned_df.to_csv('cleaned_players.csv', index=False)

print("CSV file cleaned and saved as 'cleaned_players.csv'.")
````

````python
import csv
from bs4 import BeautifulSoup

def convert_ipl_data_to_csv(html_content, output_file):
    """
    Convert IPL players data from HTML to CSV format
    
    Args:
        html_content (str): HTML content containing the IPL players data
        output_file (str): Name of the output CSV file
    """
    # Parse HTML content
    soup = BeautifulSoup(html_content, 'html.parser')
    
    # Find all team containers
    team_containers = soup.find_all('div', class_='AUipl_lst-wrp')
    
    # Prepare data for CSV
    players_data = []
    
    for team_container in team_containers:
        # Get team name
        team_name = team_container.find('h2', class_='AUipl_ttl').text.strip()
        
        # Find all player rows in the current team
        player_rows = team_container.find_all('tr', class_='AUipl_tbl-tr')
        
        # Skip header row
        for row in player_rows[1:]:  # Skip the header row
            columns = row.find_all('td')
            if columns:
                # Extract player name
                player_name = row.find('span', class_='AUipl_tbl-nm').text.strip()
                
                # Extract player category
                category = row.find('div', class_='AUipl_tbl-DXty').text.strip()
                
                # Extract price
                price = row.find('span', class_='AUipl_tx-CtRt').text.strip()
                
                # Add to players data
                players_data.append([
                    player_name,
                    team_name,
                    category,
                    price
                ])
    
    # Write to CSV file
    with open(output_file, 'w', newline='', encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile)
        # Write header
        writer.writerow(['Player', 'Team', 'Category', 'Price (₹ Cr.)'])
        # Write player data
        writer.writerows(players_data)

# Example usage
html_content = """

""" 

with open('sold_last_year.html','r') as f:
    html_content = f.read()


# Save to CSV
convert_ipl_data_to_csv(html_content, 'ipl_players.csv')
````
